## (OLD) Number of pauli strings and depth overhead for H4 and H6 for original H and fully commuting

```python
# Linear H4
import tequila as tq
from tequila.grouping.compile_groups import compile_commuting_parts 

mol = tq.Molecule(geometry="h 0.0 0.0 0.0\nh 0.0 0.0 1.5\nh 0.0 0.0 3.0\nh 0.0 0.0 4.5", basis_set="sto-3g").use_native_orbitals()

H = mol.make_hamiltonian()
print(f"len(H): {len(H)}")

options = {
    'method': 'si',  # or any other method like 'lf', 'rlf', 'ics'
    'condition': 'qwc',
    'optimize':'yes'
}
fc_groups_and_unitaries, sample_ratios = compile_commuting_parts(H, unitary_circuit='improved', options=options)
number_fc_groups = len(fc_groups_and_unitaries)
print("QWC")
print('The number of groups to measure is: ', number_fc_groups)
E = tq.ExpectationValue(H=H, U=tq.QCircuit(), optimize_measurements=options)
md = 0
for e in E.get_expectationvalues():
    depth = tq.compile_circuit(e.U).depth
    # print(depth)
    if depth > md: md = depth
print("depth overhead: ", md)

options2 = {
    'method': 'si',  # or any other method like 'lf', 'rlf', 'ics'
    'condition': 'fc',  # fully commuting condition
    'optimize':'yes'
}
fc_groups_and_unitaries, sample_ratios = compile_commuting_parts(H, unitary_circuit='improved', options=options2)
number_fc_groups = len(fc_groups_and_unitaries)
print("FC")
print('The number of groups to measure is: ', number_fc_groups)
E = tq.ExpectationValue(H=H, U=tq.QCircuit(), optimize_measurements=options2)
md = 0
for e in E.get_expectationvalues():
    depth = tq.compile_circuit(e.U).depth
    # print(depth)
    if depth > md: md = depth
print("depth overhead: ", md)
```

```python
# gin H6
import tequila as tq
from tequila.grouping.compile_groups import compile_commuting_parts 
import pandas as pd

df = pd.read_csv("h6_100")
df = df.iloc[:,10:]

original = []
qwc = []
overhead_qwc = []
fully_commuting = []
overhead_fc = []

for k in range(20):

    print(f"Molecule n{k}")

    # Define geometry
    geometry = f"""
H {df["x_0"][k]} {df["y_0"][k]} {df["z_0"][k]}
H {df["x_1"][k]} {df["y_1"][k]} {df["z_1"][k]}
H {df["x_2"][k]} {df["y_2"][k]} {df["z_2"][k]}
H {df["x_3"][k]} {df["y_3"][k]} {df["z_3"][k]}
H {df["x_4"][k]} {df["y_4"][k]} {df["z_4"][k]}
H {df["x_5"][k]} {df["y_5"][k]} {df["z_5"][k]}
"""

    mol = tq.Molecule(geometry=geometry, basis_set="sto-3g").use_native_orbitals()
    H = mol.make_hamiltonian()
    print(f"len(H): {len(H)}")
    original.append(len(H))

    options = {
    'method': 'si',  # or any other method like 'lf', 'rlf', 'ics'
    'condition': 'qwc',
    'optimize':'yes'
    }
    fc_groups_and_unitaries, sample_ratios = compile_commuting_parts(H, unitary_circuit='improved', options=options)
    number_fc_groups = len(fc_groups_and_unitaries)
    print("QWC")
    print('The number of groups to measure is: ', number_fc_groups)
    qwc.append(number_fc_groups)
    E = tq.ExpectationValue(H=H, U=tq.QCircuit(), optimize_measurements=options)
    md = 0
    for e in E.get_expectationvalues():
        depth = tq.compile_circuit(e.U).depth
        # print(depth)
        if depth > md: md = depth
    print("depth overhead: ", md)
    overhead_qwc.append(md)

    options2 = {
        'method': 'si',  # or any other method like 'lf', 'rlf', 'ics'
        'condition': 'fc',  # fully commuting condition
        'optimize':'yes'
    }
    fc_groups_and_unitaries, sample_ratios = compile_commuting_parts(H, unitary_circuit='improved', options=options2)
    number_fc_groups = len(fc_groups_and_unitaries)
    print("FC")
    print('The number of groups to measure is: ', number_fc_groups)
    fully_commuting.append(number_fc_groups)
    E = tq.ExpectationValue(H=H, U=tq.QCircuit(), optimize_measurements=options2)
    md = 0
    for e in E.get_expectationvalues():
        depth = tq.compile_circuit(e.U).depth
        # print(depth)
        if depth > md: md = depth
    print("depth overhead: ", md)
    overhead_fc.append(md)
```
