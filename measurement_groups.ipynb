```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Linear H4", "Linear H6", "Free H6", "Linear H8")
penguin_means = {
    'LF': (28, 90, 74, 154),
    'RLF': (19, 60, 54, 111),
    'SI': (19, 68, 64, 114),
    'Fermionic-based': (10, 22, 20, 33),
    'Scenario I': (9, 9, 45, 6),
    'Scenario II': (9, 9, 12, 6)
}

penguin_errors = {
    'LF': (0, 0, 15, 0),
    'RLF': (0, 0, 8, 0),
    'SI': (0, 0, 9, 0),
    'Fermionic-based': (0, 0, 2, 0),
    'Scenario I': (0, 0, 38, 0),
    'Scenario II': (0, 0, 20, 0)
}

x = np.arange(len(species))  # the label locations
width = 0.14  # the width of the bars
multiplier = 0.5

fig, ax = plt.subplots(layout='constrained')
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    
    # Plot bars without any error bars
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    
    # Add error bars only for "Free H6" (index 2)
    if penguin_errors[attribute]:
        # Only for the Free H6 position (index 2)
        ax.errorbar(x[2] + offset, measurement[2], 
                   yerr=penguin_errors[attribute][2], 
                   fmt='none', ecolor='black', capsize=5)
    
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of measurement groups')
# ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend()
ax.set_ylim(0, 170)
plt.savefig('filename.png', dpi=300)

plt.show()
```
